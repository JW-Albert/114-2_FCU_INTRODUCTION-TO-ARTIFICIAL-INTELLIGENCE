```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from keras.utils import to_categorical
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense
from ipywidgets import interact_manual
```

```python
(x_train_image, y_train_label), (x_test_image, y_test_label) = mnist.load_data()
```

```python
x_Train = x_train_image.reshape(60000, 784).astype('float32')
x_Test = x_test_image.reshape(10000, 784).astype('float32')
```

```python
x_Train_normalize = x_Train / 255
x_Test_normalize = x_Test / 255
```

```python
y_Train_OneHot = to_categorical(y_train_label)
y_Test_OneHot = to_categorical(y_test_label)
```

```python
model = Sequential()
```

```python
model.add(Dense(units=256, input_dim=784, kernel_initializer='normal', activation='relu'))
```

```python
model.add(Dense(units=10, kernel_initializer='normal', activation='softmax'))
```

```python
print(model.summary())
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
train_history = model.fit(x=x_Train_normalize, y=y_Train_OneHot, validation_split=0.2, epochs=10, batch_size=200, verbose=0)
```

```python
scores = model.evaluate(x_Test_normalize, y_Test_OneHot, batch_size=1)
print('accuracy=',scores[1])
```

```python
pre = model.predict(x_Test_normalize, batch_size=1)
prediction = np.argmax(pre, axis=1)
prediction
```
