```python
def calculate(option ,x ,y):
  match option:
    case 1:
      print("{} + {} = {}".format(x ,y ,x + y))
    case 2:
      print("{} - {} = {}".format(x ,y ,x - y))
    case 3:
      print("{} * {} = {}".format(x ,y ,x * y))
    case 4:
      print("{} / {} = {}".format(x ,y ,x / y))
    case _:
      print("Selected operation fail")
```

```python
print("Select operation:")
print("1. Add")
print("2. Subtract")
print("3. Multiply")
print("4. Divide")
```

```python
choice = 0
choice = int(input("Enter choice(1/2/3/4): "))
```

```python
first = 0
first = int(input("Enter first number: "))
```

```python
second = 0
second = int(input("Enter second number: "))
```

```python
def main():
  calculate(choice ,first ,second)
```
