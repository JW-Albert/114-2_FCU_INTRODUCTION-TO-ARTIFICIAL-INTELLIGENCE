```python
x = 0
y = 0
```

```python
x = int(input("請輸入 x (1~9): "))
y = int(input("請輸入 y (1~9): "))
if x<=0 or x>=10 or y<=0 or y>=10:
  print("範圍錯誤")
  exit()
```

```python
for i in range(1 ,x+1 ,1):
  for j in range(1 ,y+1 ,1):
    print("{} * {} = {:>2d}".format(i ,j ,i*j) ,end="\t")
  print()
```
