```python
# LR
import numpy as np
from sklearn import datasets ,linear_model
import matplotlib.pyplot as plt
import random
```

```python
upper_bound = 100
lower_bound = 5
numbers = 10
```

```python
# Make two datasets
x = []
y = []

# Put random numbers to x and y asix' list
for i in range(numbers):
  x.append(random.randint(lower_bound, upper_bound))
  y.append(random.randint(lower_bound, upper_bound))

x_arr = np.array(x).reshape(-1, 1) # Reshapes to (samples, 1 feature)
y_arr = np.array(y)
```

```python
# Training a LR model
LR_model = linear_model.LinearRegression()
LR_model.fit(x_arr ,y_arr)
```

```python
# Put origianl x back to the model
y_pre = LR_model.predict(x_arr)
```

```python
plt.scatter(x_arr ,y_arr ,color='black')
plt.plot(x_arr ,y_pre ,color='blue' ,linewidth=3)

plt.xlabel("X")
plt.ylabel("Y")

plt.show()
```

```python
delta = []
for i in range(numbers):
  x_new = LR_model.predict([x_arr[i]])
  delta.append(abs(x[i] - x_new))
```

```python
total_delta = sum(delta)
print("Tolerance =" ,total_delta[0])
```
