## Import Package

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np
```

```python
def random_KMeans(x_min ,x_max ,y_min ,y_max ,num_samples ,k):
  x = np.random.uniform(x_min ,x_max ,num_samples)
  y = np.random.uniform(y_min ,y_max ,num_samples)

  # 將 x, y 組合建構成 K-Means 需要的資料格式 (N, 2)
  data_arr = np.column_stack((x ,y))

  clf = KMeans(n_clusters=k)
  clf.fit(data_arr)

  plt.figure(figsize=(8 ,6))
  plt.scatter(data_arr[:, 0], data_arr[:, 1], s=100, c=clf.labels_)
  plt.title(f"K-Means Clustering (k={k})")
  plt.show()
```

```python
random_KMeans(20, 80, 20, 80, 20, 2)
```
